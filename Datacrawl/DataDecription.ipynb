```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
pd.options.display.float_format = '{:,.5f}'.format
def datasetAdvanceDescrible(Idataset):
  stats = Idataset.describe()
  stats.loc['var'] = Idataset.var().tolist()
  stats.loc['skew'] = Idataset.skew().tolist()
  stats.loc['kurt'] = Idataset.kurtosis().tolist()
  return stats
```

```python
# 2. Đọc file csv
# company_csv = '/content/drive/MyDrive/Colab Notebooks/Datacrawl/IMP.csv'
# company_csv = '/content/drive/MyDrive/Colab Notebooks/Datacrawl/OPC.csv'
# company_csv = '/content/drive/MyDrive/Colab Notebooks/Datacrawl/VMD.csv'

dfIMP = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datacrawl/IMP.csv')
dfIMP
# và gắng index với giá Close cho df1
dfIMPclose = dfIMP.reset_index()['close']
dfIMPclose
```

```python
print(datasetAdvanceDescrible(dfIMPclose))
```

```python
dfVMD = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datacrawl/VMD.csv')
dfVMD
# và gắng index với giá Close cho df1
dfVMDclose = dfVMD.reset_index()['close']
dfVMDclose
```

```python
print(datasetAdvanceDescrible(dfVMDclose))
```

```python
dfOPC = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datacrawl/OPC.csv')
dfOPC
# và gắng index với giá Close cho df1
dfOPCclose = dfOPC.reset_index()['close']
dfOPCclose
```
